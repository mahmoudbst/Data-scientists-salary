# Job Title

```python
def job_title(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    if 'machine learning' in title.lower() or 'mle' in title.lower():
        return 'mle'
    if 'data engineer' in title.lower():
        return 'data engineer'
    if 'analyst' in title.lower():
        return 'analyst'
    if 'manager' in title.lower():
        return 'manager'
    if 'director' in title.lower():
        return 'director'
    else:
        return 'na'
    
```

```python
df['Title'] = df['Job Title'].apply(lambda x: job_title(x))
```

```python
df.Title.value_counts()
```

# Salary Cleaning

```python
df['Salary Estimate'].value_counts()
```

```python
df.info()
```

```python
df['Salary Estimate'].value_counts()
```

```python
df=df[df['Salary Estimate'] !='-1']
```

```python
df['Salary Estimate'].unique()
```

```python
def clean_salary_estimate(data):
    data=data.replace('$','')
    data=data.replace('K','')
    data=data.replace('(Glassdoor est.)','')
    data=data.lower().replace('per hour','')
    data=data.lower().replace('employer provided salary:','')
    data=data.lower().replace('(employer est.)','')
    return(data)
    
    
```

```python
def get_min(data):
    return data.split('-')[0]
```

```python
def get_max(data):
    return data.split('-')[1]
```

```python
df['Hourly']=df['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
```

```python
df['Employer_Provided_Salary']=df['Salary Estimate'].apply(lambda x: 1 if 'employer provided salary' in x.lower() else 0)
```

```python
df
```

```python
df.Employer_Provided_Salary.value_counts()
```

```python
df.Hourly.value_counts()
```

```python
salary=df['Salary Estimate'].apply(lambda x:clean_salary_estimate(x))
```

```python
salary.unique()
```

```python
df['Min Salary(K)']=salary.apply(lambda x:int(get_min(x)))
```

```python
df['Max Salary(K)']=salary.apply(lambda x:int(get_max(x)))
```

```python
df.dtypes
```

Converting Hourly salaries to annual salaries and considering an average of 2000 hours a year

```python
df['Min Salary(K)']=df.apply(lambda x: x['Min Salary(K)']*2 if x.Hourly==1 else x['Min Salary(K)'],axis=1)
df['Max Salary(K)']=df.apply(lambda x: x['Max Salary(K)']*2 if x.Hourly==1 else x['Max Salary(K)'],axis=1)
```

```python
df['Avg Salary(K)']=(df['Max Salary(K)']+df['Min Salary(K)'])/2
```

# Seniority

```python
def seniority(data):
    if 'senior' in data.lower() or 'sr.' in data.lower() or 'sr' in data.lower() or 'lead' in data.lower() or 'principal' in data.lower():
        return 'senior'
    if 'jr' in data.lower() or 'jr.' in data.lower() or 'junior' in data.lower() or 'entry' in data.lower():
        return 'junior'
    else:
        return 'na'
    
```

```python
df['Seniority']=df['Job Title'].apply(lambda x:seniority(x))
```

```python
df['Seniority'].value_counts()
```
