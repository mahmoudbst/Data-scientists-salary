# Exploring categorical features

```python
df.columns
```

```python
df_cat = df[['Location', 'Headquarters', 'Size','Type of ownership', 'Industry', 'Sector', 'Revenue','Company_name_txt', 'State', 'SQL','Python', 'Seniority']]
```

```python
for i in df_cat.columns:
    cat=df_cat[i].value_counts()[:30]
    print('graph for %s : total =%d' % (i,len(df_cat[i].value_counts())))
    chart = sns.barplot(x = cat.index, y = cat)
    chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
    plt.show()
```
